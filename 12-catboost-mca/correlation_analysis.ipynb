```python
import sickle
import seaborn
from matplotlib import pyplot as plt
```

```python
%matplotlib inline
```

```python
CUR_DIR, ROOT = sickle.paths('asdfas')
```

```python
sc01 = sickle.load_sc_scanpy('SC01', 'SC01v2')
```

```python
preds = sickle.load_predictions('./sc01-preds.csv', 'MCA')
```

```python
preds.drop(preds.index[~preds.index.isin(sc01.obs.index)], inplace=True)
```

```python
cell_type = preds.idxmax(axis=1)
pred_score = preds.max(axis=1)
```

```python
def violin(df, x, y, title, xticklabels=None):
    plt.figure(figsize=(14, 8))
    seaborn.stripplot(x=x, y=y, data=df, jitter=True, alpha=0.4, color="black", size=2)
    ax = seaborn.violinplot(x=x, y=y, data=df, inner=None)
    if xticklabels:
        ax.set_xticklabels(xticklabels)
    ax.tick_params(axis='x', labelrotation=90)
    ax.set_title(title)
    plt.tight_layout()
    ax.figure.savefig('./sc01-{}.png'.format(y))
    return ax
```

```python
df = sc01.obs.copy()
```

```python
df['prediction'] = cell_type[df.index]
df['pred_score'] = pred_score[df.index]
```

```python
violin(df, 'prediction', 'percent_mito', 'Percent of mito genes by predicted cluster')
```

```python
violin(df, 'prediction', 'percent_ribo', 'Percent of ribo genes by predicted cluster')
```

```python
violin(df, 'prediction', 'n_genes', 'Number of genes by predicted cluster')
```

```python
violin(df, 'prediction', 'n_UMI', 'Number of UMIs by predicted cluster')
```

```python
violin(df, 'prediction', 'pred_score', 'Prediction confidence by predicted cluster')
```
